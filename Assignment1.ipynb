```python
import pandas as pd
import numpy as np
import pickle

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#conda install scikit-learn
```

```python
data = ("Assign1_Housing data.csv")
df = pd.read_csv(data)
df
```

```python
df.info() 
# non null

```

```python
df.replace({"yes": 1, "no": 0}, inplace=True)
print(df)
```

```python
df.replace({"furnished": 2, "semi-furnished": 1, "unfurnished":0 }, inplace = True)
print(df)
```

```python
X = df.iloc[ : ,1:]
y = df.iloc[ : ,0]

X_train, X_test, y_train, y_test = train_test_split(X,y ,test_size = 0.2, random_state = 79)

```

```python
linear_model = LinearRegression().fit(X_train,y_train)

predict = linear_model.predict(X_test)
```

```python
mse = mean_squared_error(y_test,predict)
print(f'MSE: {mse:.2f}')
```

```python

import joblib

# save model with joblib
filename = 'ml_6610422015.sav'
joblib.dump(linear_model, filename)
```
